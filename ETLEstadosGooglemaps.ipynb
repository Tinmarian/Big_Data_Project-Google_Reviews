```python
import pandas as pd
from IPython.display import display

# Cargar datos desde un archivo CSV en tu computadora local

archivo_csv = "C:/Users/Fede/Documents/Henry/Proyecto 3/Pruebas de Código en local/ReviewsEstados.csv"
ReviewsEstados = pd.read_csv(archivo_csv)

# Mostrar el contenido del DataFrame usando display
display(ReviewsEstados)
```

```python
# Calcular el porcentaje de valores nulos por columna
porcentaje_nulos = (ReviewsEstados.isnull().sum() / len(ReviewsEstados)) * 100

# Crear un DataFrame para mostrar los resultados
df_porcentaje_nulos = pd.DataFrame({'Columna': ReviewsEstados.columns, 'Porcentaje de Nulos': porcentaje_nulos})

# Mostrar el porcentaje de valores nulos para cada columna específica
columnas_interes = ['user_id', 'name', 'time', 'rating', 'text', 'pics', 'resp', 'gmap_id']
porcentaje_nulos_columnas = df_porcentaje_nulos[df_porcentaje_nulos['Columna'].isin(columnas_interes)]

# Mostrar el contenido del DataFrame de porcentaje de nulos para las columnas de interés usando display
display(porcentaje_nulos_columnas)
```

```python
# Eliminar las columnas 'pics' y 'resp' del DataFrame ReviewsEstados
ReviewsEstados = ReviewsEstados.drop(['pics', 'resp'], axis=1)

# Mostrar el DataFrame resultante después de eliminar las columnas
display(ReviewsEstados)
```

```python
# Detectar filas duplicadas
duplicados = ReviewsEstados.duplicated()

# Eliminar filas duplicadas
ReviewsEstados = ReviewsEstados.drop_duplicates()
```

```python
# Calcular estadísticas descriptivas, incluyendo percentiles
estadisticas = ReviewsEstados.describe()

# Calcular el rango intercuartílico (IQR)
Q1 = ReviewsEstados.quantile(0.25)
Q3 = ReviewsEstados.quantile(0.75)
IQR = Q3 - Q1

# Identificar y eliminar valores atípicos basados en IQR
ReviewsEstados = ReviewsEstados[~((ReviewsEstados < (Q1 - 1.5 * IQR)) | (ReviewsEstados > (Q3 + 1.5 * IQR))).any(axis=1)]

# Mostrar el DataFrame resultante
display(ReviewsEstados)
```

```python
# Guardar el DataFrame en un archivo CSV
ReviewsEstados.to_csv('ReviewEstadosFinal.csv', index=False)
```
