```python
import pandas as pd
from IPython.display import display
import glob

# Lista de rutas a los archivos JSON
rutas = [
    "C:/Users/Fede/Documents/Henry/Proyecto 3/Google maps/Metadatos sitio/1.json",
    "C:/Users/Fede/Documents/Henry/Proyecto 3/Google maps/Metadatos sitio/2.json",
    "C:/Users/Fede/Documents/Henry/Proyecto 3/Google maps/Metadatos sitio/3.json",
]

# Lista para almacenar los DataFrames
dataframes = []

for ruta in rutas:
    df = pd.read_json(ruta, lines=True)
    dataframes.append(df)

# Concatenar los DataFrames en uno solo
Metadatos_Final = pd.concat(dataframes, ignore_index=True)

# Mostrar el DataFrame concatenado
display(Metadatos_Final)

# Guardar el DataFrame concatenado en un archivo CSV llamado "Metadatos-Final.csv"
Metadatos_Final.to_csv("Metadatos-Final1.csv", index=False)
```
