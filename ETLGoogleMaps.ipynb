```python
import pandas as pd
from IPython.display import display

# Cargar el primer archivo CSV
df_metadata = pd.read_csv("C:/Users/Fede/Documents/Henry/Proyecto 3/etl/df_metadata1.csv")

# Cargar el segundo archivo CSV y renombrar la columna "name" como "user_name"
df_reviews = pd.read_csv("C:/Users/Fede/Documents/Henry/Proyecto 3/Pruebas de Código en local/ReviewsEstados.csv")
df_reviews.rename(columns={'name': 'user_name'}, inplace=True)

# Concatenar las filas por "gmap_id"
GoogleMapsDF = pd.concat([df_metadata, df_reviews], axis=0, ignore_index=True, sort=False)

# Guardar el DataFrame resultante en un archivo CSV
GoogleMapsDF.to_csv("GoogleMapsDF.csv", index=False)

# Mostrar el DataFrame resultante usando display
display(GoogleMapsDF)
```
