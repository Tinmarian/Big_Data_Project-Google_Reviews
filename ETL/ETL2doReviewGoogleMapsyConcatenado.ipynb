```python
import pandas as pd

# Ruta del archivo CSV de entrada
archivo_entrada = 'C:/Users/Fede/Documents/Henry/Proyecto 3/etl/GoogleMapsCompletoDF_Concatenado2.csv'

# Cargar el archivo CSV en un DataFrame
df = pd.read_csv(archivo_entrada)

# Eliminar la columna 'name'
df.drop(columns=['name'], inplace=True)

# Cambiar el nombre de la columna 'name_x' a 'name'
df.rename(columns={'name_x': 'name'}, inplace=True)

# Cambiar los tipos de las columnas según las especificaciones
df['name'] = df['name'].astype(str)
df['address'] = df['address'].astype(str)
df['category'] = df['category'].astype(str)
df['hours'] = df['hours'].astype(str)
df['MISC'] = df['MISC'].astype(str)
df['state'] = df['state'].astype(str)
df['user_id'] = df['user_id'].astype(str)
df['time'] = pd.to_datetime(df['time'])
df['text'] = df['text'].astype(str)


# Guardar el DataFrame modificado en un nuevo archivo CSV
archivo_salida = 'GoogleMapsCompletoDF_Concatenado3.csv'
df.to_csv(archivo_salida, index=False)

# Mostrar el DataFrame con display
display(df)
```
