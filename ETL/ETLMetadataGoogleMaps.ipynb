Función para quitar emojis

```python
def remove_emojis(text):
    return demoji.replace(text, '')
```

```python
def remove_emojis_2(text):
    if type(text) == str:
        return demoji.replace(text, '')
    else:
        return None 
```

ESTADOS: Función para unir todos los archivos en un solo dataframe

```python
def crear_df(estado,num_archivos):
    df_1 = pd.DataFrame()
    for i in range(1, num_archivos):
        df = pd.read_json(f'../Google Maps/reviews-estados/review-{estado}/{i}.json', lines=True)
        df_1 = pd.concat([df_1, df], ignore_index=True)

    return df
```

METADATA: Función para unir todos los archivos en un solo dataframe

```python
def crear_df_metadata(num_archivos):
    df_1 = pd.DataFrame()
    for i in range(1, num_archivos):
        df = pd.read_json(f'../Google Maps/metadata-sitios/{i}.json', lines=True)
        df_1 = pd.concat([df_1, df], ignore_index=True)

    return df
```
