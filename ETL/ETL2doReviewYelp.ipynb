```python
import pandas as pd

# Cargar el archivo CSV
archivo_csv = "C:/Users/Fede/Documents/Henry/ProyectoFinal/Yelp/YelpReviewBusinessLimpio.csv"
df = pd.read_csv(archivo_csv)

# Lista de nombres a filtrar, con posibles variaciones en mayÃºsculas/minÃºsculas y errores tipogrÃ¡ficos
nombres_a_filtrar = ["starbucks", "mcdonald's", "kfc", "subway", "domino's pizza"]

# Filtrar los nombres en la columna 'name' (ignorando mayÃºsculas/minÃºsculas y errores tipogrÃ¡ficos)
filtro = df['name'].str.lower().str.replace('[^\w\s]', '').isin(nombres_a_filtrar)
resultados_filtrados = df[filtro]

# Mostrar los resultados con display
display(resultados_filtrados)

# Guardar los resultados en un archivo CSV
resultados_filtrados.to_csv("Restaurants.csv", index=False)
```

```python
import pandas as pd

# Cargar el archivo CSV
archivo_csv = "C:/Users/Fede/Documents/Henry/ProyectoFinal/Yelp/Restaurants.csv"
df = pd.read_csv(archivo_csv)

# Crear una columna de fecha combinando las columnas "anio," "mes," y "dia"
df['date'] = pd.to_datetime(df[['anio', 'mes', 'dia']])

# Ordenar el DataFrame por la columna 'date' de menor a mayor
df_ordenado = df.sort_values(by='date')

# Mostrar los resultados con display
display(df_ordenado)

# Guardar los resultados en un archivo CSV
archivo_salida = "C:/Users/Fede/Documents/Henry/ProyectoFinal/Yelp/RestaurantsOrdenadoxLocalxFecha.csv"
df_ordenado.to_csv(archivo_salida, index=False)
```
