```python
import pandas as pd

# Ruta al archivo CSV original
archivo_csv = 'C:/Users/Fede/Documents/Henry/Proyecto 3/etl/Business2.csv'

# Cargar el archivo CSV en un DataFrame
df = pd.read_csv(archivo_csv)

# Transformar las columnas según las especificaciones
df['stars'] = df['stars'].astype(float)
df['review_count'] = df['review_count'].astype(int)
df.drop(columns=['postal_code', 'latitude', 'longitude', 'is_open'], inplace=True)

# Guardar el DataFrame resultante en un nuevo archivo CSV
df.to_csv('Business3.csv', index=False)

# Mostrar el DataFrame resultante con display
pd.set_option('display.max_columns', None)  # Mostrar todas las columnas
display(df)
```

```python
import pandas as pd

# Rutas a los archivos CSV
archivo_business = 'C:/Users/Fede/Documents/Henry/Proyecto 3/etl/Business3.csv'
archivo_review = 'C:/Users/Fede/Documents/Henry/Proyecto 3/etl/ReviewYelpConcat.csv'

# Cargar ambos archivos CSV en DataFrames
df_business = pd.read_csv(archivo_business)
df_review = pd.read_csv(archivo_review)

# Realizar la concatenación usando la columna "business_id"
df_resultado = pd.merge(df_review, df_business, on='business_id', how='inner')

# Guardar el resultado en un nuevo archivo CSV
df_resultado.to_csv('YelpReviewyBusiness.csv', index=False)

# Mostrar el DataFrame resultante con display
pd.set_option('display.max_columns', None)  # Mostrar todas las columnas
display(df_resultado)
```
