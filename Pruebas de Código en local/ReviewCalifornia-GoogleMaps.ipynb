```python
import pandas as pd
import json
import os

# Lista de rutas de archivos JSON
rutas_archivos = [
    "C:/Users/Fede/Documents/Henry/Proyecto 3/Google maps/Reviews-estados/review-California/1.json",
    "C:/Users/Fede/Documents/Henry/Proyecto 3/Google maps/Reviews-estados/review-California/2.json",
    "C:/Users/Fede/Documents/Henry/Proyecto 3/Google maps/Reviews-estados/review-California/3.json",
    "C:/Users/Fede/Documents/Henry/Proyecto 3/Google maps/Reviews-estados/review-California/4.json",
    "C:/Users/Fede/Documents/Henry/Proyecto 3/Google maps/Reviews-estados/review-California/5.json",
    "C:/Users/Fede/Documents/Henry/Proyecto 3/Google maps/Reviews-estados/review-California/6.json",
    "C:/Users/Fede/Documents/Henry/Proyecto 3/Google maps/Reviews-estados/review-California/7.json",
    "C:/Users/Fede/Documents/Henry/Proyecto 3/Google maps/Reviews-estados/review-California/8.json",
    "C:/Users/Fede/Documents/Henry/Proyecto 3/Google maps/Reviews-estados/review-California/9.json",
    "C:/Users/Fede/Documents/Henry/Proyecto 3/Google maps/Reviews-estados/review-California/10.json",
    "C:/Users/Fede/Documents/Henry/Proyecto 3/Google maps/Reviews-estados/review-California/11.json",
    "C:/Users/Fede/Documents/Henry/Proyecto 3/Google maps/Reviews-estados/review-California/12.json",
    "C:/Users/Fede/Documents/Henry/Proyecto 3/Google maps/Reviews-estados/review-California/13.json",
    "C:/Users/Fede/Documents/Henry/Proyecto 3/Google maps/Reviews-estados/review-California/14.json",
    "C:/Users/Fede/Documents/Henry/Proyecto 3/Google maps/Reviews-estados/review-California/15.json",
    "C:/Users/Fede/Documents/Henry/Proyecto 3/Google maps/Reviews-estados/review-California/16.json",
    "C:/Users/Fede/Documents/Henry/Proyecto 3/Google maps/Reviews-estados/review-California/17.json",
    "C:/Users/Fede/Documents/Henry/Proyecto 3/Google maps/Reviews-estados/review-California/18.json"
]

# Crear una lista para almacenar los datos JSON
datos_json = []

# Recorrer todas las rutas de archivos y cargar los datos JSON
for ruta_archivo in rutas_archivos:
    try:
        with open(ruta_archivo, 'r', encoding='utf-8') as archivo:
            # Leer líneas del archivo
            lineas = archivo.readlines()

        for linea in lineas:
            # Intentar cargar cada línea como un objeto JSON y agregarlo a la lista
            try:
                datos_json.append(json.loads(linea))
            except json.JSONDecodeError as e:
                print(f"Error al decodificar la línea como JSON en {ruta_archivo}: {e}")
    except FileNotFoundError:
        print(f"El archivo JSON no se encontró en la ruta especificada: {ruta_archivo}")
    except Exception as e:
        print(f"Ocurrió un error: {e}")

# Crear un DataFrame a partir de la lista de datos JSON
df_review_california_final = pd.DataFrame(datos_json)

# Mostrar el DataFrame
from IPython.display import display
display(df_review_california_final)

#Guardar el DataFrame en un archivo CSV
df_review_california_final.to_csv("review-California-Final.csv", index=False)
```
