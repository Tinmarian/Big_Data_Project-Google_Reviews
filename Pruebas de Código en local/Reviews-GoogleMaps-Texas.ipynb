```python
import pandas as pd
import json
import os

# Lista de rutas de archivos JSON
rutas_archivos = [
    "C:/Users/Fede/Documents/Henry/Proyecto 3/Google maps/Reviews-estados/review-Texas/1.json",
    "C:/Users/Fede/Documents/Henry/Proyecto 3/Google maps/Reviews-estados/review-Texas/2.json",
    "C:/Users/Fede/Documents/Henry/Proyecto 3/Google maps/Reviews-estados/review-Texas/3.json",
    "C:/Users/Fede/Documents/Henry/Proyecto 3/Google maps/Reviews-estados/review-Texas/4.json",
    "C:/Users/Fede/Documents/Henry/Proyecto 3/Google maps/Reviews-estados/review-Texas/5.json",
    "C:/Users/Fede/Documents/Henry/Proyecto 3/Google maps/Reviews-estados/review-Texas/6.json",
    "C:/Users/Fede/Documents/Henry/Proyecto 3/Google maps/Reviews-estados/review-Texas/7.json",
    "C:/Users/Fede/Documents/Henry/Proyecto 3/Google maps/Reviews-estados/review-Texas/8.json",
    "C:/Users/Fede/Documents/Henry/Proyecto 3/Google maps/Reviews-estados/review-Texas/9.json",
    "C:/Users/Fede/Documents/Henry/Proyecto 3/Google maps/Reviews-estados/review-Texas/10.json",
    "C:/Users/Fede/Documents/Henry/Proyecto 3/Google maps/Reviews-estados/review-Texas/11.json",
    "C:/Users/Fede/Documents/Henry/Proyecto 3/Google maps/Reviews-estados/review-Texas/12.json",
    "C:/Users/Fede/Documents/Henry/Proyecto 3/Google maps/Reviews-estados/review-Texas/13.json",
    "C:/Users/Fede/Documents/Henry/Proyecto 3/Google maps/Reviews-estados/review-Texas/14.json",
    "C:/Users/Fede/Documents/Henry/Proyecto 3/Google maps/Reviews-estados/review-Texas/15.json",
    "C:/Users/Fede/Documents/Henry/Proyecto 3/Google maps/Reviews-estados/review-Texas/16.json",
]

# Crear una lista para almacenar los datos JSON
datos_json = []

# Recorrer todas las rutas de archivos y cargar los datos JSON
for ruta_archivo in rutas_archivos:
    try:
        with open(ruta_archivo, 'r', encoding='utf-8') as archivo:
            # Leer lÃ­neas del archivo
            lineas = archivo.readlines()

        for linea in lineas:
            # Intentar cargar cada lÃ­nea como un objeto JSON y agregarlo a la lista
            try:
                datos_json.append(json.loads(linea))
            except json.JSONDecodeError as e:
                print(f"Error al decodificar la lÃ­nea como JSON en {ruta_archivo}: {e}")
    except FileNotFoundError:
        print(f"El archivo JSON no se encontrÃ³ en la ruta especificada: {ruta_archivo}")
    except Exception as e:
        print(f"OcurriÃ³ un error: {e}")

# Crear un DataFrame a partir de la lista de datos JSON
df_review_New_York_final = pd.DataFrame(datos_json)

# Mostrar el DataFrame
from IPython.display import display
display(df_review_New_York_final)

#Guardar el DataFrame en un archivo CSV
df_review_New_York_final.to_csv("review-New_York-Final.csv", index=False)
```
