```python
import pandas as pd
from IPython.display import display  # Importa la funci√≥n display

# Ruta del archivo "Business.csv"
ruta_del_archivo_csv = "C:/Users/Fede/Documents/Henry/Proyecto 3/Business.csv"

# Cargar el archivo "Business.csv" en un DataFrame
data = pd.read_csv(ruta_del_archivo_csv)

# Nombres a filtrar
nombres_a_filtrar = ['KFC', "Domino's Pizza", 'McDonald\'s', 'Subway', 'Pizza Hut']

nombres_a_filtrar = [nombre.lower() for nombre in nombres_a_filtrar]

# Filtrar las filas que contienen los nombres especificados (insensible a may√∫sculas y min√∫sculas)
filas_filtradas = data[data['name'].str.lower().isin(nombres_a_filtrar)]

# Mostrar el DataFrame filtrado con display
display(filas_filtradas)

# Calcular el promedio de estrellas para cada nombre filtrado
promedio_estrellas = filas_filtradas.groupby('name')['stars'].mean()

# Mostrar el DataFrame con los promedios utilizando display
display(promedio_estrellas.reset_index())
```

```python
import pandas as pd
from IPython.display import display  # Importa la funci√≥n display

# Ruta del archivo "Business.csv"
ruta_del_archivo_csv = "C:/Users/Fede/Documents/Henry/Proyecto 3/Business.csv"

# Cargar el archivo "Business.csv" en un DataFrame
data = pd.read_csv(ruta_del_archivo_csv)

# Nombres a filtrar
nombres_a_filtrar = ['KFC', "Domino's Pizza", 'McDonald\'s', 'Subway', 'Pizza Hut']

# Convertir los nombres a min√∫sculas para que la comparaci√≥n sea insensible a may√∫sculas y min√∫sculas
nombres_a_filtrar = [nombre.lower() for nombre in nombres_a_filtrar]

# Filtrar las filas que contienen los nombres especificados (insensible a may√∫sculas y min√∫sculas)
filas_filtradas = data[data['name'].str.lower().isin(nombres_a_filtrar)]

# Realizar el conteo de cada nombre
conteo_nombres = filas_filtradas['name'].value_counts()

# Mostrar el conteo de nombres con display
display(conteo_nombres)

# Guardar el DataFrame en un archivo CSV llamado "business.csv"
data.to_csv("business.csv", index=False)
```
